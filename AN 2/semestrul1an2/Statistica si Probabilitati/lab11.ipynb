```python
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt


def f(x):
    return np.exp(-1*x**2)

def simuleaza_uniforma(a, b):
    return np.random.uniform(low = a, high=b)

def simuleaza_exp(n, a, b, const=1):
    vals = []
    
    for i in range(n):
        u = simuleaza_uniforma(a, b)
        x = f(u)
        vals.append(x)
        
    return np.mean(vals)*const

    
a = 0
b = 5
n = 100
const = 5
rasp = simuleaza_exp(n=n, a=a, b=b, const=const)
print(rasp)
        
    
```

```python
a = 0
b = 5
n = 1000
const = 5

def deseneaza_convergenta(nr_tr=1, ep=0.5, rasp = 0.88):
    for i in range(nr_tr):
        simulari = []
        for j in range(n):
            val = simuleaza_exp(j, a, b, const = const)
            simulari.append(val)
        plt.plot(simulari)
    plt.axhline(y=ep + rasp, color = "red")
    plt.axhline(y=rasp - ep, color = "red")
    plt.show()

deseneaza_convergenta(nr_tr = 5, ep = 0.3)
    
    
```

```python
from scipy.stats import expon

def deseneaza_teorema(n=30, nr_samples = 100):
    
    rezultat = []
    for i in range(1, n + 1):
        
        vals = expon.rvs(loc = 0, scale = 0.1, size = (i, nr_samples))
        #vals = np.random.uniform(size=(i,nr_samples))
        vals = np.sum(vals, axis=0)
        rezultat.append(vals)
    return rezultat
        
rez = deseneaza_teorema(n=100, nr_samples=1000)

for i, simulare in enumerate(rez):
    #if i > 70:
    fig = plt.hist(simulare, bins = 30)
    plt.show()
        
```
