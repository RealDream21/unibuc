```python
import numpy as np
import matplotlib.pyplot as plt
import scipy

from scipy.stats import expon
```

```python
#convergenta in distributie

def fn(n, x):
    return 1 - (1 - 1/n)**(n*x)
    
domeniu = np.linspace(0, 5, 1000)
val_exp = [expon.cdf(x) for x in domeniu]

plt.plot(domeniu, val_exp)


def deseneaza_conv_in_dist(domeniu, ns=5):
    for i in range(1, ns + 1):
        val_i = [fn(i, x) for x in domeniu]
        plt.plot(domeniu, val_i)
        
    return
        
    
deseneaza_conv_in_dist(domeniu, 5)
plt.show()
    

```
