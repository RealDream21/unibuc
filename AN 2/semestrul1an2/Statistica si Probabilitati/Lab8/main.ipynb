```python
#librarii: numpy, scipy
#simulam o binomiala
#matplotlib -> deseneaza chestii
#scipy -> binomial -> 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#incercam sa simulam o binomiala
n = 10
p = 0.9
samples = np.random.binomial(n=n, p=p, size=10000)

plt.hist(samples, density = True)
```

```python
#Incercam sa simulam o geometrica
#numpy.random.geometric
geom_p = 0.6
samples = np.random.geometric(p = geom_p, size = 1000)
plt.hist(samples)
```

```python
#distributia poisson
#numpy.random.poisson
size = 100000
lam = 3
lam_samples = np.random.poisson(lam = lam, size = size)

N = 30
p = 0.1
bin_samples = np.random.binomial(n = N, p = p, size = size)

fig, ax = plt.subplots(1, 2)
ax[0].hist(lam_samples)
ax[1].hist(bin_samples)
plt.show()

plt.hist(lam_samples, alpha = 0.5)
plt.hist(bin_samples, alpha = 0.5)
plt.show()
```
