```python
import scipy
import numpy as np
import matplotlib.pyplot as plt

expon = scipy.stats.expon

lam = 1
scale = 1/lam
```

```python
simulari = expon.rvs(scale=scale, size= 1000) #rvs = random variates

domeniu = np.linspace(0, 10)
val_densitate = expon.pdf(domeniu)

plt.hist(simulari, density=True)
plt.plot(domeniu, val_densitate)
```

```python
norm = scipy.stats.norm

scale = 1 #scale aka standard deviation
loc = 0

simulari = norm.rvs(loc=loc, scale=scale, size= 1000)
domeniu = np.linspace(-4, 4)
val_densitate = norm.pdf(domeniu)


plt.hist(simulari, alpha=0.6, density=True)
plt.hist(simulari, val_densitate)

```
