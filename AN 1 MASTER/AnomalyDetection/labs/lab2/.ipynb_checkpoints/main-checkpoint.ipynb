```python
#pyod.utils.data.generate_data_clusters -> gen clusters of synth data
#sklearn.datasets.make_blobs -> clustere cu matrice d-aia
#scipy.io.loadmat -> loads matrix from matlab
#pyod.utils.utility.standardizer -> standardizare medie 0 dev std 1
#pyod.models.combination.average -> average of outlier scores
#pyod.models.combination.maximization -> takes maximum score
#numpy.quantile -> calculeaza a q a cuantila dupa o axa si un input
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Random1DLiniarModelGenerator():
    mu_x: float
    sigma_x: float
    mu_y: float
    sigma_y: float
    a: int
    b: int

    def __init__(self, mu_x, sigma_x, mu_y, sigma_y, a, b):
        self.mu_x = mu_x
        self.sigma_x = sigma_x
        self.mu_y = mu_y
        self.sigma_y = sigma_y
        self.a = a
        self.b = b
        
    def generate_sample(self):
        epsilon = np.random.normal(loc=self.mu_y, scale=self.sigma_y**2)
        x = np.random.normal(loc=self.mu_x, scale=self.sigma_x) 
        return (x, self.a * x + self.b + epsilon)

    def generate_n_samples(self, n):
        samples = np.array([self.generate_sample() for _ in range(n)])
        return samples
        
    
```

```python
generator_low_noise = Random1DLiniarModelGenerator(mu_x = 0, sigma_x = 1, mu_y = 0, sigma_y = 1, a = 2, b = 3)
generate_high_variance_x = Random1DLiniarModelGenerator(mu_x = 0, sigma_x = 5, mu_y = 0, sigma_y = 1, a = 2, b = 3)
generate_high_variance_y = Random1DLiniarModelGenerator(mu_x = 0, sigma_x = 1, mu_y = 0, sigma_y = 5, a = 2, b = 3)
generate_high_variance_x_y = Random1DLiniarModelGenerator(mu_x = 0, sigma_x = 5, mu_y = 0, sigma_y = 5, a = 2 , b = 3)

n_samples = 100
low_noise_points = generator_low_noise.generate_n_samples(n = n_samples)
high_variance_x_points = generate_high_variance_x.generate_n_samples(n = n_samples)
high_variance_y_points = generate_high_variance_y.generate_n_samples(n = n_samples)
high_variance_x_y_points = generate_high_variance_x_y.generate_n_samples(n = n_samples)
```

```python
fig, plots = plt.subplots(2, 2, figsize=(10, 10))
plots.ravel()
plots[0, 0].scatter(low_noise_points[:, 0], low_noise_points[:, 1])
plots[0, 0].set_title("Low noise on x and y")
plots[0, 1].scatter(high_variance_x_points[:, 0], high_variance_x_points[:, 1])
plots[0, 1].set_title("High variance on x, low variance on y")
plots[1, 0].scatter(high_variance_y_points[:, 0], high_variance_y_points[:, 1])
plots[1, 0].set_title("Low variance on x, high variance on y")
plots[1, 1].scatter(high_variance_x_y_points[:, 0], high_variance_x_y_points[:, 1])
plots[1, 1].set_title("High variance on x, high variance on y")

for x, y in zip([0, 0, 1, 1], [0, 1, 0, 1]):
    plots[x, y].set_xlim(-10, 10)
    plots[x, y].set_ylim(-60, 100)

plt.show()
```

```python
def get_leverage_scores(X):
    U, S, Vh = np.linalg.svd(a=X, full_matrices=False)
    return np.sum(U**2, axis = 1)

def get_highest_leverage_points(X):
    scores = get_leverage_scores(X)
    threshold = np.quantile(scores, 0.95)
    return scores >= threshold

def get_colors_points(X):
    labels = get_highest_leverage_points(X)
    return ['b' if x == False else 'r' for x in labels]
```

```python
fig, plots = plt.subplots(2, 2, figsize=(10, 10))
plots.ravel()
plots[0, 0].scatter(low_noise_points[:, 0], low_noise_points[:, 1], c=get_colors_points(low_noise_points))
plots[0, 0].set_title("Low noise on x and y")
plots[0, 1].scatter(high_variance_x_points[:, 0], high_variance_x_points[:, 1], c=get_colors_points(high_variance_x_points))
plots[0, 1].set_title("High variance on x, low variance on y")
plots[1, 0].scatter(high_variance_y_points[:, 0], high_variance_y_points[:, 1], c=get_colors_points(high_variance_y_points))
plots[1, 0].set_title("Low variance on x, high variance on y")
plots[1, 1].scatter(high_variance_x_y_points[:, 0], high_variance_x_y_points[:, 1], c = get_colors_points(high_variance_x_y_points))
plots[1, 1].set_title("High variance on x, high variance on y")

for x, y in zip([0, 0, 1, 1], [0, 1, 0, 1]):
    plots[x, y].set_xlim(-20, 20)
    plots[x, y].set_ylim(-60, 100)
plots[0, 0].set_xlim(-5, 5)
plots[0, 0].set_ylim(-10, 10)

plt.show()
```

```python
class Random2DLiniarModelGenerator():
    mu_x1: float
    sigma_x1: float
    mu_x2: float
    sigma_x2: float
    mu_y: float
    sigma_y: float
    a: int
    b: int
    c: int

    def __init__(self, mu_x1, sigma_x1, mu_x2, sigma_x2, mu_y, sigma_y, a, b, c):
        self.mu_x1 = mu_x1
        self.sigma_x1 = sigma_x1
        self.mu_x2 = mu_x2
        self.sigma_x2 = sigma_x2
        self.mu_y = mu_y
        self.sigma_y = sigma_y
        self.a = a
        self.b = b
        self.c = c
        
    def generate_sample(self):
        epsilon = np.random.normal(loc=self.mu_y, scale=self.sigma_y**2)
        x1 = np.random.normal(loc=self.mu_x1, scale=self.sigma_x1)
        x2 = np.random.normal(loc=self.mu_x2, scale=self.sigma_x2)
        return (x1, x2, self.a * x1 + self.b * x2 + self.c + epsilon)

    def generate_n_samples(self, n):
        samples = np.array([self.generate_sample() for _ in range(n)])
        return samples
```

```python
low_mean = 0
low_var = 1
high_var = 5
a = 2
b = 3
c = 5
n_samples = 100
low_noise_points = Random2DLiniarModelGenerator(low_mean, low_var, low_mean, low_var, low_mean, low_var, a, b, c).generate_n_samples(n_samples)
high_noise_x1_points = Random2DLiniarModelGenerator(low_mean, high_var, low_mean, low_var, low_mean, low_var, a, b, c).generate_n_samples(n_samples)
high_noise_x2_points = Random2DLiniarModelGenerator(low_mean, low_var, low_mean, high_var, low_mean, low_var, a, b, c).generate_n_samples(n_samples)
high_noise_y_points = Random2DLiniarModelGenerator(low_mean, low_var, low_mean, low_var, low_mean, high_var, a, b, c).generate_n_samples(n_samples)
high_noise_x1_x2_y_points = Random2DLiniarModelGenerator(low_mean, high_var, low_mean, high_var, low_mean, high_var, a, b, c).generate_n_samples(n_samples)
```

```python
fig, plots = plt.subplots(2, 3, figsize=(10, 10), subplot_kw={'projection': '3d'})

plots[0, 0].scatter(low_noise_points[:, 0], low_noise_points[:, 1], low_noise_points[:, 2],
                    c=get_colors_points(low_noise_points))
plots[0, 0].set_title("Low noise")
plots[0, 1].scatter(high_noise_x1_points[:, 0], high_noise_x1_points[:, 1], high_noise_x1_points[:, 2],
                    c=get_colors_points(high_noise_x1_points))
plots[0, 1].set_title("High noise on x1")
plots[0, 2].scatter(high_noise_x2_points[:, 0], high_noise_x2_points[:, 1], high_noise_x2_points[:, 2],
                    c=get_colors_points(high_noise_x2_points))
plots[0, 2].set_title("High noise on x2")
plots[1, 0].scatter(high_noise_y_points[:, 0], high_noise_y_points[:, 1], high_noise_y_points[:, 2],
                    c=get_colors_points(high_noise_y_points))
plots[1, 0].set_title("High noise on y")
plots[1, 1].scatter(high_noise_x1_x2_y_points[:, 0], high_noise_x1_x2_y_points[:, 1], high_noise_x1_x2_y_points[:, 2],
                    c=get_colors_points(high_noise_x1_x2_y_points))
plots[1, 1].set_title("High noise on x1, x2, and y")

plots[1, 2].axis('off')
plt.subplots_adjust(hspace=0.0, wspace=0.2)
plt.show()
```

```python
from pyod.utils.data import generate_data_clusters
from pyod.models.knn import KNN
from sklearn.metrics import confusion_matrix
```

```python
X_train, X_test, y_train, y_test = generate_data_clusters(n_train=400, n_test=200, n_clusters=2, contamination=0.1)
neighbors = [1, 3, 5, 7, 9, 11, 13]

for n_neighbors in neighbors:
    knn = KNN(contamination=0.1, n_neighbors=n_neighbors)
    y_train_preds = knn.fit_predict(X_train)
    y_test_preds = knn.predict(X_test)

    fig, plots = plt.subplots(2, 2, figsize=(10, 10))
    
    plots[0, 0].scatter(X_train[:, 0], X_train[:, 1], c = ['b' if x == 0 else 'r' for x in y_train])
    plots[0, 0].set_title(f"Training data ground truth, n_neighbors={n_neighbors}")
    plots[0, 1].scatter(X_train[:, 0], X_train[:, 1], c = ['b' if x == 0 else 'r' for x in y_train_preds])
    plots[0, 1].set_title(f"Training data predictions, n_neighbors={n_neighbors}")
    plots[1, 0].scatter(X_test[:, 0], X_test[:, 1], c = ['b' if x == 0 else 'r' for x in y_test])
    plots[1, 0].set_title(f"Test data ground truth, n_neighbors={n_neighbors}")
    plots[1, 1].scatter(X_test[:, 0], X_test[:, 1], c = ['b' if x == 0 else 'r' for x in y_test_preds])
    plots[1, 1].set_title(f"Test data predictions, n_neighbors={n_neighbors}")

    cm_train = confusion_matrix(y_train, y_train_preds)
    cm_test = confusion_matrix(y_test, y_test_preds)

    TPR_train = (cm_train[1, 1]) / (cm_train[1, 1] + cm_train[1, 0])
    TNR_train = (cm_train[0, 0]) / (cm_train[0, 0] + cm_train[0, 1])

    TPR_test = (cm_test[1, 1]) / (cm_test[1, 1] + cm_test[1, 0])
    TNR_test = (cm_test[0, 0]) / (cm_test[0, 0] + cm_test[0, 1])

    print(f"Balanced accuracy train: {(TPR_train + TNR_train)/ 2}")
    print(f"Balanced accuracy test:{(TPR_test + TNR_test) / 2}")
    
    
    plt.show()
```

```python
from sklearn.datasets import make_blobs
from pyod.models.lof import LOF
```

```python
cluster_1, _ = make_blobs(n_samples=200, centers=[[-10, -10]], cluster_std=2)
cluster_2, _ = make_blobs(n_samples=100, centers=[[10, 10]], cluster_std=6)
data = np.concatenate((cluster_1, cluster_2))
```

```python
neighbors = [1, 3, 5, 7, 9, 11, 13]

for n_neighbors in neighbors:
    knn = KNN(contamination=0.07, n_neighbors=n_neighbors)
    lof = LOF(contamination=0.07, n_neighbors=n_neighbors)
    preds_knn = knn.fit_predict(data)
    preds_lof = lof.fit_predict(data)

    fig, plots = plt.subplots(1, 2, figsize=(18, 9))

    plots[0].scatter(data[:, 0], data[:, 1], c = ['r' if x == 1 else 'b' for x in preds_knn])
    plots[0].set_title(f"KNN, k={n_neighbors}")
    plots[1].scatter(data[:, 0], data[:, 1], c = ['r' if x == 1 else 'b' for x in preds_lof])
    plots[1].set_title(f"LOF, k={n_neighbors}")
    plt.show()
```
